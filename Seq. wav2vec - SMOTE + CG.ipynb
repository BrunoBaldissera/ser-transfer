```python
X_cetuc_train, X_cetuc_test, y_cetuc_train, y_cetuc_test = train_test_split(X_cetuc, y_cetuc, test_size=0.2, random_state=42)
print(X_cetuc_train.shape, X_cetuc_test.shape, y_cetuc_train.shape, y_cetuc_test.shape)
```

```python
import numpy as np
from keras import Model, layers
from keras.layers import Dense, Input
from matplotlib import pyplot
from sklearn.metrics import classification_report

shared_layer = Dense(30, kernel_initializer='normal', activation='relu', name='shared_layer')

#cetuc
input_layer_cetuc = Input(shape=(44,), name='cetuc_input')
layer_1_cetuc = shared_layer(input_layer_cetuc)
output_layer_cetuc = Dense(1, kernel_initializer='normal', activation="sigmoid", name='cetuc_output')(layer_1_cetuc)
model_cetuc = Model(inputs=[input_layer_cetuc], outputs=[output_layer_cetuc])
model_cetuc.compile(loss='mean_squared_error', optimizer='adam')
model_cetuc.fit(X_cetuc_train, y_cetuc_train, epochs=100, batch_size=1000, validation_data=(X_cetuc_test, y_cetuc_test), verbose=2, shuffle=False)

print("------------ start ser -------------")

from sklearn.metrics import  f1_score
# K-fold Cross Validation model evaluation
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, shuffle=True)

fold_no = 1
f1_per_fold = []
acc_per_fold = []
loss_per_fold = []
for train, test in kfold.split(X_SER, y_SER):
  X_ser_train_824 = X_SER[train, :-44]
  X_ser_train_44 = X_SER[train, -44:]
  X_ser_test_824 = X_SER[test, :-44]
  X_ser_test_44 = X_SER[test, -44:]
  print(train)
  y_ser_train = y_SER[train]
  y_ser_test = y_SER[test]

  input_layer_ser_1 = Input(shape=(768,), name='ser_input_1')
  input_layer_ser_2 = Input(shape=(44,), name='ser_input_2')

  layer_1_ser_1 = Dense(10, kernel_initializer='normal', activation='relu', name='layer_1_ser_1')(input_layer_ser_1)
  layer_1_ser_2 = Dense(10, kernel_initializer='normal', activation='relu', name='layer_1_ser_2')(input_layer_ser_2)

  #shared congelada
  shared_layer_ser = shared_layer(input_layer_ser_2)
  shared_layer_ser.trainable = False

  #predição task 1 (M / F)
  mf_predict_ser = Dense(1, kernel_initializer='normal', activation="relu", name='mf_predict_ser')(shared_layer_ser)

  merged_ser_1 = layers.concatenate([layer_1_ser_1, layer_1_ser_2])
  merged_ser = layers.concatenate([merged_ser_1, mf_predict_ser])

  # 5 -> 10
  layer_2_ser = Dense(10, kernel_initializer='normal', activation='relu', name='layer_2_ser')(input_layer_ser_1)

  output_layer_ser = Dense(3, kernel_initializer='normal', activation="relu", name='ser_output')(layer_2_ser)

  model = Model(inputs=[input_layer_ser_1, input_layer_ser_2], outputs=output_layer_ser)
  model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])


  # Fit data to model
  history = model.fit([X_ser_train_824, X_ser_train_44], y_ser_train, epochs=300, batch_size=100, validation_data=([X_ser_test_824, X_ser_test_44], y_ser_test), verbose=1, shuffle=False)
 
  #fscore
  y_pred = model.predict([X_ser_test_824, X_ser_test_44], batch_size=64, verbose=1)
  y_pred_c = np.argmax(y_pred, axis=1)
  y_test_c = np.argmax(y_ser_test, axis=1)
  print(classification_report(y_test_c, y_pred_c, zero_division=0))
  _val_f1 = f1_score(y_test_c, y_pred_c, average='macro', zero_division=0)
  print ("val_f1: ", _val_f1)
  f1_per_fold.append(_val_f1)

  scores = model.evaluate([X_ser_test_824, X_ser_test_44], y_ser_test, verbose=1)
  print(model.metrics_names, scores)
  print(f'Score for fold {fold_no}: {model.metrics_names[0]} of {scores[0]}; {model.metrics_names[1]} of {scores[1]*100}%')
  acc_per_fold.append(scores[1] * 100)
  loss_per_fold.append(scores[0])

  
  # Increase fold number
  fold_no = fold_no + 1

# == Provide average scores ==
print('------------------------------------------------------------------------')
print('Score per fold')
for i in range(0, len(acc_per_fold)):
  print('------------------------------------------------------------------------')
  print(f'> Fold {i+1} - F1-Macro: {f1_per_fold[i]} - Loss: {loss_per_fold[i]} - Accuracy: {acc_per_fold[i]}%')
print('------------------------------------------------------------------------')
print('Average scores for all folds:')
print(f'> F1-Macro: {np.mean(f1_per_fold)} (+- {np.std(f1_per_fold)})')
print(f'> Accuracy: {np.mean(acc_per_fold)} (+- {np.std(acc_per_fold)})')
print(f'> Loss: {np.mean(loss_per_fold)}')
print('------------------------------------------------------------------------')
```
